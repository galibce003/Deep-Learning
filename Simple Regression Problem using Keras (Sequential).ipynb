## Split into explanatory and response variable

```python
X = pd.DataFrame(df.iloc[:, 0:1])
y = pd.DataFrame(df.iloc[:, 1:2])
```

<br/>

## Split into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 1)
```

```python
X_train.shape
```

<br/>

## Sequential model build up through Keras
model.add(Dense(400.... - There will be 400 neurons in a single layer  
input_dim = 1 - At a time there will entry one variable only  
<br/>

model.add(Dense(1...... - As we will be predicting single value, in the last layer, there will be only one neuron.  
activation='linear' - For a regression problem, the last activation will always be 'linear'.  

```python
model = Sequential()
model.add(Dense(400, input_dim=1, activation='relu'))
model.add(Dense(200, input_dim=200, activation='relu'))
model.add(Dense(200, input_dim=200, activation='relu'))
model.add(Dense(1, activation='linear'))
```

<br/>

## Now time to train our data as well as model

```python
keras.optimizers.Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, amsgrad=False)
model.compile(loss='mean_squared_error', optimizer='RMSprop', metrics=['mean_absolute_percentage_error'])
```

## Model summary
from the summary we can see that there are three dense layers and the last layer is our expected output layer.

```python
model.summary()
```

<br/>

## Fitting the model

```python
history = model.fit(X_train, y_train, epochs=500, batch_size=32,validation_split=0.15,validation_data=None,verbose=1)
```

<br/>

## Clear Session
If we want to change something regarding model layers, first we will have to clear the previous session. Only after that the change will be worked out.

```python
keras.backend.clear_session()
```

## Prediction

```python
y_pred = model.predict(X_test)
```

<br/>

## Creating a pandas dataframe with the actual and predicted values

```python
dfff = pd.DataFrame(np.c_[y_test, y_pred], columns = ['Actual', 'Predicted'])
```

```python
dfff.head()
```

## Plot the actual and predicted values

```python
plt.plot(dfff['Actual'], color = 'Black', label = 'Actual')
plt.plot(dfff['Predicted'], color = 'Red', label = 'Predicted', alpha = 0.6)
plt.legend()

plt.show()
```
